# Load Dependencies

```python
import matplotlib.pyplot as plt
```

# Load Data

```python
data1_x=['0-20','20-40','40-60','60-80','80-100']
data1_steric_y=[4031,617,104,0,0]
data1_electronic_y=[1548,1425,1394,286,99]
data2_x=['0.0-0.2','0.2-0.4','0.4-0.6','0.6-0.8','0.8-1.0']
data2_steric_y=[1611,1252,989,144,4]
data2_electronic_y=[3611,353,36,0,0]
```

# Results of Data 1 

## Change of Predicted Yield(%) by Eliminating Steric Encodings


```python
fig = plt.figure(figsize=(8,4))
tick_fontsize = 15
label_fontsize = 18
plt.bar(list(range(len(data1_x))),data1_steric_y,color='lightcoral',width=0.56)
plt.xticks(list(range(len(data1_x))),
           data1_x,
          fontsize=tick_fontsize,rotation=30)
plt.yticks(fontsize=tick_fontsize)
plt.ylabel('Count',fontsize=label_fontsize)
print()
```

## Change of Predicted Yield(%) by Eliminating Electronic Encodings


```python
fig = plt.figure(figsize=(8,4))
tick_fontsize = 15
label_fontsize = 18
plt.bar(list(range(len(data1_x))),data1_electronic_y,color='lightsteelblue',width=0.56)
plt.xticks(list(range(len(data1_x))),
           data1_x,
          fontsize=tick_fontsize,rotation=30)
plt.yticks(fontsize=tick_fontsize)
plt.ylabel('Count',fontsize=label_fontsize)
print()
```
